```python
# Getting Started.
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
# Retorna números espaçados igualmente dado o intervalo.
# De 0 até 2 * o número pi, 200 amostras.
y = np.sin(x)
fig, ax = plt.subplots()

ax.plot(x, y)
plt.show()
```

```python
# plot(x, y)
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

x = np.linspace(0, 10, 200)
# Divide o intervalo de 0 até 10 por duzentos e retorna tudo isso.
# Vai retornar 200 amostras que vão de 0 até 10.
y = 4 + 2 * np.sin(2 * x)

# plot
fig, axes = plt.subplots()

# axes: eixo
axes.plot(x, y, linewidth=2.0)

axes.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
