```python
# Getting Started.
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
# Retorna números espaçados igualmente dado o intervalo.
# De 0 até 2 * o número pi, 200 amostras.
y = np.sin(x)
fig, ax = plt.subplots()

ax.plot(x, y)
plt.show()
```

```python
np.random.seed(19680801)
data = {
    'a': np.arange(50),
    'c': np.random.randint(0, 50, 50), # vai de 0 a 50, 50 números assim.
    'd': np.random.randn(50) # Retorna amostras da distribuição normal padrão.
}

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

ax.scatter('a', 'b', c = 'c', s = 'd', data = data)
ax.set_xlabel('Entrada A')
ax.set_ylabel('Entrada B')
```

```python
def my_plotter(ax, data1, data2, param_dict):
    """
        Uma função que faz gráfico.
    """
    out = ax.plot(data1, data2, **param_dict)
    return out

data1, data2, data3, data4 = np.random.randn(4, 100)

fix, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})

```

```python
# Cada axes(que é tipo o gráfico) pode ter 2 ou 3 objetos Axis(o eixo no gráfico) representanto o x e o y.
# Eles controlam a escala do Axis.

# Escala
fig, axs = plt.subplots(1, 2, figsize=(5, 2.7), layout='constrained')
xdata = np.arange(len(data1))
data = 10 ** data1
axs[0].plot(xdata, data)

axs[1].set_yscale('log') # Mudei a escala pra log
axs[1].plot(xdata, data)
```

```python
#  Dados mapeados por cor.

X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

fig, axs = plt.subplots(2, 2, layout='constrained', figsize=(6, 4))
pc = axs[0, 0].pcolormesh(X, Y, Z, vmin=-1, vmax=1, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[0, 0])
axs[0, 0].set_title('pcolormesh()')

co = axs[0, 1].contourf(X, Y, Z, levels=np.linspace(-1.25, 1.25, 11))
fig.colorbar(co, ax=axs[0, 1])
axs[0, 1].set_title('contourf()')

pc = axs[1, 0].imshow(Z ** 2 * 100, cmap = 'plasma', norm = mpl.colors.LogNorm(vmin=0.01, vmax=100))

fig.colorbar(pc,ax=axs[1,0], extend='both')
axs[1, 1].scatter(data1, data2, c = data3, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[1, 1], extend='both')
axs[1, 1].set_title('scatter()')
```
